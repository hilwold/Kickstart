```python
ts1in = open("Data/ts1_input.txt","r")
ts1inline = ts1in.readline()
T = int(ts1inline)
ts1out = open("Data/ts1_output.txt","r")


for loop in range(T):
    L = int(ts1in.readline())
    S = ts1in.readline()
    print(S)
    count = 0
    ans = f"Case #{loop + 1}:"
    for i in range(L):
        #print('i :',i, S[i])
        if count == 0 : 
            #print(count,end=' ')
            count = 1
        elif ( count > 0):
            if (ord(S[i - 1]) < ord(S[i])):
                #print(count,end=' ')
                count += 1
            else:
                count = 1
        #print('',count,end="")
        
        ans  = ans + ' ' + str(count)
    ans = ans + '\n'
    ts1outline = ts1out.readline()
    if ans == ts1outline:
        print('Pass')
    else:
        print('Fail')
    
        
```
